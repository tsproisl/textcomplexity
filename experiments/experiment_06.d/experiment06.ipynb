# Standardized type-token ratio: Mean vs. boundaries of confidence interval

```python
%matplotlib inline

import matplotlib.pyplot as plt
import pandas as pd
import scipy
import seaborn as sns

plt.rcParams['figure.figsize'] = (12.0, 8.0)
```

```python
data = pd.read_csv("sttr_ci.tsv", sep="\t", header=0, index_col=0)
```

```python
data_for_boxplots = data.copy()
data_for_boxplots["brow_genre"] = data_for_boxplots["genre"]
low_brow = data[data.genre != "hlit"].copy()
low_brow["brow_genre"] = "low"
data_for_boxplots = pd.concat((data_for_boxplots, low_brow))
```

```python
sns.countplot(x="brow_genre", data=data_for_boxplots)
plt.rcParams['figure.figsize'] = (12.0, 8.0)
```

## Standardized type-token ratio

```python
sns.boxplot(x="brow_genre", y="sttr", data=data_for_boxplots)
```

## Lower boundary of confidence interval

```python
data_for_boxplots["sttr_lower"] = data_for_boxplots["sttr"] - data_for_boxplots["ci"]
sns.boxplot(x="brow_genre", y="sttr_lower", data=data_for_boxplots)
```

## Upper boundary for confidence interval

```python
data_for_boxplots["sttr_upper"] = data_for_boxplots["sttr"] + data_for_boxplots["ci"]
sns.boxplot(x="brow_genre", y="sttr_upper", data=data_for_boxplots)
```
